# Encoding Categorical attributes

```python
y = pd.get_dummies(data.Species, prefix='Species')
data['Species_Bream']=y['Species_Bream']
data['Species_Parkki']=y['Species_Parkki']
data['Species_Perch']=y['Species_Perch']
data['Species_Smelt']=y['Species_Smelt']
data['Species_Pike']=y['Species_Pike']
data['Species_Roach']=y['Species_Roach']
data['Species_Whitefish']=y['Species_Whitefish']
data
```

```python
data.drop(['Species'], axis=1, inplace=True)
data
```

# Normalizing data


```python
minmax=MinMaxScaler()

cols=list(data.columns)
cols

for i in range(len(cols)):
    data[cols[i]]=pd.DataFrame(minmax.fit_transform(data[[cols[i]]]))
```

# Displaying min/max value of every column

```python
for i in range(len(cols)):
    print(cols[i], 'Min:', data[cols[i]].min(),'Max:',data[cols[i]].max())
```
